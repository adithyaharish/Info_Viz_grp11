```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Agriculture_raw_2016.csv')
df.head(10)
```

```python
df.columns
```

```python
df_new = df[['Program', 'Year','State','Commodity', 'Data Item','Value']]
df_new
```

```python
df_new['Data Item'] = df_new['Data Item'].str.split('-').str[1]
df_new
```

```python
df_new['Commodity'].value_counts()
```

```python
df_new['Data Item'].value_counts()
```

```python
df_new.columns
```

```python
df_new.info()
```

```python
df_new['Value'] = df_new['Value'].replace(to_replace=r'\s*\(D\)\s*', value='0', regex=True)
df_new['Value'] = df_new['Value'].replace(to_replace=r'\s*\(NA\)\s*', value='0', regex=True)
```

```python
df_new['Value'] = df_new['Value'].str.replace(',', '').astype(int)
df_new['Value'] = df_new['Value'].astype(int)
```

```python
df_pivoted = df_new.pivot_table(index=['Program', 'Year', 'State', 'Commodity'],
                                  columns='Data Item', values='Value', fill_value=0).reset_index()

# Flatten the columns (optional step for column renaming)
df_pivoted.columns = [col if not isinstance(col, tuple) else col[1] for col in df_pivoted.columns.values]

print(df_pivoted)
```

```python
df_pivoted
```

```python
df_pivoted.to_csv('Agriculture_final_2016.csv', index=False)
```
