```python
import pandas as pd

# Read the agriculture data
df = pd.read_csv('Agriculture_final_2016.csv')

# Filter the dataset for California
california_data = df[df['State'] == 'CALIFORNIA']

# Calculate the total production cost for all commodities in California
total_production_cost_all = california_data[' PRODUCTION, MEASURED IN $'].sum()

# Filter the dataset for barley in California
barley_data = california_data[california_data['Commodity'] == 'BARLEY']

# Calculate the total production cost for barley in California
total_production_cost_barley = barley_data[' PRODUCTION, MEASURED IN $'].sum()

# Display the results
print("Total Production Cost for All Commodities in California:", total_production_cost_all)
print("Total Production Cost for Barley in California:", total_production_cost_barley)
```

```python
df1['Station.State'] = df1['Station.State'].replace({'DE': 'Delaware', 'VA': 'Virginia'})

# Calculate average temperatures by state and year
average_temperatures = df1[(df1['Date.Year'] == 2016) & (df1['Station.State'] != 'Puerto Rico')].groupby(['Station.State', 'Date.Year'])['Data.Temperature.Avg Temp'].mean().reset_index()
average_temperatures.columns = ['State', 'Year', 'Average Temperature']
average_temperatures
```

```python
# Calculate total production by state, commodity, and year
sum_production = df.groupby(['State', 'Commodity']).agg({' PRODUCTION, MEASURED IN $': 'sum'}).reset_index()
sum_production.columns = ['State', 'Commodity', 'Production']
```

```python
# Merge dataframes on the State column
sum_production['State'] = sum_production['State'].str.title()
combined_df = pd.merge(sum_production, average_temperatures, on='State')
```

```python
# Map state names to abbreviations if using abbreviations
state_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL',
    'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL',
    'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY',
    'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA',
    'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH',
    'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC',
    'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR',
    'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT',
    'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV',
    'Wisconsin': 'WI', 'Wyoming': 'WY'
}
combined_df['State'] = combined_df['State'].map(state_abbrev)
```

```python
# Create dropdown widget
all_commodities = ['All'] + list(df['Commodity'].unique())
dropdown = widgets.Dropdown(options=all_commodities, value='All', description='Commodity:')

# Create output widget
output = Output()

# Define function to update choropleth map based on selected commodity
def update_map(commodity):
    if commodity == 'All':
        # Calculate total production cost for all commodities in each state
        combined_df['Total_Production_Cost'] = combined_df.groupby('State')['Production'].transform('sum')
        # Create hover text with total production cost for all commodities in each state
        combined_df['Hover_Text'] = combined_df.apply(lambda x: f"Commodities: {', '.join(combined_df[combined_df['State'] == x['State']]['Commodity'].unique())}<br>Total Production Cost: ${x['Total_Production_Cost']:,}<br>Average Temperature: {x['Average Temperature']:.2f}°F", axis=1)
       
        # Create a choropleth map for 'All' option
        fig = px.choropleth(combined_df,
                            locations="State",  
                            locationmode='USA-states',  
                            scope="usa",  
                            color="Production",  
                            hover_name="State",  
                            hover_data=["Hover_Text"],  # Additional data to show on hover
                            title="Average Crop Yield and Temperature by State",
                            labels={"Production": "Total Crop Yield"}
                           )
    else:
        # Calculate total production cost for the selected commodity in each state
        filtered_df = combined_df[combined_df['Commodity'].str.contains(commodity)]
        filtered_df['Total_Production_Cost'] = filtered_df.groupby('State')['Production'].transform('sum')
        # Create hover text with total production cost for the selected commodity in each state
        filtered_df['Hover_Text'] = filtered_df.apply(lambda x: f"Commodity: {commodity}<br>Total Production Cost: ${x['Total_Production_Cost']:,}<br>Average Temperature: {x['Average Temperature']:.2f}°F", axis=1)
       
        # Create a choropleth map for the selected commodity
        fig = px.choropleth(filtered_df,
                            locations="State",  
                            locationmode='USA-states',  
                            scope="usa",  
                            color="Production",  
                            hover_name="State",  
                            hover_data=["Hover_Text"],  # Additional data to show on hover
                            title=f"Total {commodity} Yield and Temperature by State",
                            labels={"Production": f"Total {commodity} Yield"}
                           )
   
    # Show the plot
    with output:
        output.clear_output()
        fig.show()
```

```python
# Connect dropdown widget to update function
widgets.interactive(update_map, commodity=dropdown)

# Display dropdown and output widget
display(dropdown, output)
```
